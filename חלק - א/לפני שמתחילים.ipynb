# כמה דברים לפני שנתחיל

היי! לפניכם המדריך על איך ליצור בוט טלגרם בפייתון

המדריך יוצא מנקודת הנחה שמי שעושה אותו מכיר את שפת פייתון לפחות ברמה התחלתית, כמו כן, מנקודת הנחה שפייתון כבר מותקן.

אני כותב את המדריך הזה, לאחר מספר רב של בקשות, כולל היום, ביום שאני מתחיל את זה, לגבי איך ליצור בוט לטלגרם.

ישנן מספר דרכים, החל מהמון ספריות שונות לטלגרם, החל משילוב של זה עם פלאסק, ועוד.

`aiogram` אנחנו הולכים להשתמש בספריה, כנראה ספריית הטלגרם הכי טובה, בשם 



#### APIלמה דווקא בה? הספריה אסינכרונית, ומטפלת לכם ברוב הדברים שאתם צריכים לטפל בהם בספריות אחרות, אפילו ברוב הקונפיגורציות מול ה

`בקיצור בגלל שזה ממש נוח וקל`

---
### !בואו נתחיל

נתחיל קודם כל, מההתקנה של הספריה שנשתמש בה, תתקינו אותה דרך הטרמינל בעזרת:

`pip install aiogram`

עוד על הספריה - https://pypi.org/project/aiogram/

במידה והתקנתם אותה,ואין תקלות...אז מגניב! מוזמנים לעבור לקובץ הבא, ששם נתחיל כבר ליצור את הבוט שלנו

במידה ומשהו לא עבד, טוב, גוגל לרשותכם

### הרצת קוד


במדריך הזה יהיו מינימום 2 קבצים כל פעם, הקובץ האחד הוא מדריך שכתוב בעזרת מחברת יופיטר עם פירוט על הקוד והשלבים

והקובץ השני יהיה קוד נקי של אותו המדריך
