# חלק ב - פקודות, טקסט, טקסט ספציפי, או מילים בתוך טקסט

בחלק הזה, אני אראה לכם מספר פונקציות, אסביר מה ההבדל ביניהן, ואפרא על כל שורה.


### פקודות

בפונקציה לפניכם, אתם יכולים לראות משהו חדש, יש בדקורטור פרמטר, הפרמטר הזה מקבל רשימה.

`commands=['עזרה', 'start']`

היא פקודה, מנגד ללא סלאש, זה יתקבל רק בתור סלאש `/start` זה בעצם אומר לבוט "לפקודות האלו, תעשה את הפונקציה הזאתי" - יש להדגיש, שפקודות שונות מטקסט, לפני כל פקודה יש סלאש, כלומר למשל 
מעבר לזה, כאשר ישנן פקודות, אז הבוט מראה אותן כשעושים סלאש, נלמד בהמשך איך לעשות את זה.





### תגובה לטקסט ספציפי

כאן, כמו בתגובות, הפונקציה הבאה תעבוד רק אם הטקסט שיוכנס, יהיה ספציפית הטקסט הזה, להבדיל מתגובות, לא יהיה אפשרי לראות אותן בתור רמיזות.






### תגובות במידה ומילה קיימת בתוך הטקסט

הפונקציה הבאה, תעבוד במידה והמילה שהגדרנו, תהיה חלק מהטקסט.
