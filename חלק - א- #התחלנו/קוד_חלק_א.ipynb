```python
# הקוד הסודי שלכן
#API_TOKEN = 'הכנס כאן'
API_TOKEN = '1482545755:AAFi4Y1opYdTSgN468iOhPWTSgB1JbVEEPU'
# כאן זה מגדיר את הבוט עצמו, ומתחבר אליו     
bot = Bot(token=API_TOKEN)
# כאן אתם מחברים את הדיספאטצ'ר לבוט, שזה בעצם מה שיפאשר לכם לתקשר עם הבוט, לעבד את מה שהוא שולח לכם, ולשלוח תשובה בחזרה, ובכללי לנהל דברים שונים
dp = Dispatcher(bot)
```

בחלק הבא, נוסיף פונקציה.
נעבור שורה שורה

`@dp.message_handler()` - דקורטור לפונקציה, בעצם הדבר הזה עוטף את הפונקציה שאנחנו משתמשים בה, ומציב לה תנאים בתנאי שרוצים, וקורם לפונקציה "להגיב" ברגע שיש פעילות מצד הבוט

`async def echo(message: types.Message):` - ככה נראית הגדרה של פונקציה אסינכרונית - חשוב לזכור שכל הספריה הנל מבוססת אסינכרוניות, שזה מה שמאפשר לה לעבוד בצורה סופר גמישה, ומאפשר מספר פונקציות עם תלויות שונות בצורה ,  
בקיצור מדהימה

**תשימו לב שבפונקציה למעלה, הפונקציה מקבלת הודעה מסוג מסויים, זה לא פרט שולי, ניהול של טקסט, קבצים, אודיו ותמונות, נעשה שם, כדי שאם למשל  יש פוקנציה כלשהי שרק היא מיועדת לתמונות, אז רק הפונקציה הזאתי תתמודד עם הבקשה שמגיע מהבט **

`await message.answer('Hello!')` - הפורמט שבו עונים לבוט - יכול להשתנות מגירסא לגירסא כמו שכבר היה.


```python
@dp.message_handler() # יו בוט, זאת פונקציה שתהיה קשורה אלייך
async def echo(message: types.Message):

    await message.answer('Hello!')
```

`if __name__ == '__main__':` - אני יוצא מנקודת הנחה שאתם יודעים פייתון, ויודעים את משמעות השורה הזאת, אבל בגדול, זאת הפונקציה העיקרית ש

`executor.start_polling(dp, skip_updates=True)` - בגדול מביא לכם ארגז עלים עצום שאפשר יכולים להשתמש בו כדי לנהל את הבוט עם מגוון דברים.

```python
if __name__ == '__main__':
    executor.start_polling(dp, skip_updates=True)
```
