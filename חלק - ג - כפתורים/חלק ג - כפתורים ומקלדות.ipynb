# חלק - ג - כפתורים, ומקלדת

בחלק הזה קורים מספר דברים.

1. יצרנו קובץ פייתון בשם keyboard 
    - בקובץ הזה משכנו קלאסים שיעזרו לנו ליצור מקלדת, או מקלדות, ועל המקלדות האלו לשים כפתורים
2. בקובץ app - 
    - אנחנו משכנו את המקלדת אלינו, והצבנו אותה בתור תשובה, מזה אפשר להבין, שאנחנו יכולים לחחזיר מספר דברים בתור תשובה.
    -  עשינו העתק הדבק לפונקציה מהחלק הקודם, והחלפנו את שלום במילה שהכנסנו לכפתור. ככה נקבל תשובה כאשר נלחץ על הכפתור

טוב אז, מה זה בעצם כפתור? הכפתור בגדול, מחזיר לכם טקסט, כלומר אם תיצרו כפתור, עם המילה "שלום" אז בעצם מה שזה יעשה, כתלחצו על הכפתור, זה ישלך בשמכם את המילה שלום לבוט.
כלומר אפשר להתאים כפתורים לפי מה שנחוץ לכם, או בתור אפשרויות עיקריות, ולבנות פונקציות בהתאם למה שנוח לכם, זה גם יקל מאוד על המשתמש.

### כמובן שאפשר להחזיר מקלדת בתור תשובה מהבוט עצמו, למשל אם נרצה שאחרי שהמשתמש יכתוב לבוט שלום, הבוט יביא למשתמש רשימה של תוספות לפיצה

`הוספנו עוד מקלדת`



`הוספנו מקלדת לתשובה, על המילה שמביא לנו הכפתור`



`תוצאה אחרי לחיצה על הכפתור שלנו`




דגשים, כנראה ראיתם בבוטים שונים, שהפתורים ממוקמים מחוץ למקדלת, יש להביל בין הכפתורים האלו, לבין הכפתורים שלנו.
מעבר לזה, לפי דעתי יהיה יותר תקין לקרוא לכפתורים שפירטתי עליהם כאן מקשים, כי בעצם הם משמשים בתור מקשים, על מקלדת.

`*  חשוב להבדיל בן המקשים לבין הכפתורים, כי שני הדברים עושים דברים מאוד שונים, ומשמשים לדברים שונים`

על כפתורים אמיתיים, נלמד בהמשך, אבל כרגע למדנו על מקשים, בחרתי להשתמש במילה כפתורים, כי לפני דעתי היחס אל הדברים האלו בתור מקשים יכל היה לבלב אנשים מתחילים.


`טיפים קטנים למקלדות`

`הקטנת הכפתורים + מקלדת חד פעמים - כלומר, המקלדת תיסגר ברתה שתלחצו על הכפתור`







`שינוי צורת המקלדת ושורות`

אפשר להרכיב מקלדות בצורות שונות, הנה אחת מהן - 
בהתחלה אנחנו יוצרים שורה של 3 כפתורים, ולאחר מכן מוסיפים כפתור בנפרד.

כמובן אפשר להוסיף עוד שורה במידה ונרצה

`row` - להוספת שורה, כל המקשים שנכניס לשם, יכנסו בתוך שורה אחת
`add` - רק מוסיף כפתורים אחרי מה שאתם שמים
